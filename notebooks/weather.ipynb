```python
import os
import json
import pandas as pd

# Initialize a list to hold the extracted data
data_list = []

# Directory where the .txt files are located
directory_path = '../data/raw/2017_2018_municipality/'

# Loop through each file in the directory
for filename in os.listdir(directory_path):
    if filename.endswith('.txt'):
        # Construct the full file path
        file_path = os.path.join(directory_path, filename)
        
        # Open and read the file line by line
        with open(file_path, 'r') as f:
            for line in f:
                data_dict = json.loads(line.strip())
                
                # Extract coordinates
                longitude, latitude = data_dict['geometry']['coordinates']
                
                # Extract relevant information from the dictionary
                data = {
                    'longitude': longitude,
                    'latitude': latitude,
                    'calculatedAt': data_dict['properties']['calculatedAt'],
                    'from': data_dict['properties']['from'],
                    'to': data_dict['properties']['to'],
                    'municipalityName': data_dict['properties']['municipalityName'],
                    'parameterId': data_dict['properties']['parameterId'],
                    'value': data_dict['properties']['value']
                }
                
                # Append the extracted data to the list
                data_list.append(data)

# Convert the list to a DataFrame
df = pd.DataFrame(data_list)
```

```python
# Minx: 11.795567590870476, Miny: 55.438414325467335, Maxx: 12.843531176748721, Maxy: 56.114397772504546
# filter data based on coordinates
df = df[df['longitude'].between(11.795567590870476, 12.843531176748721) & df['latitude'].between(55.438414325467335, 56.114397772504546)]
```

```python
# Define the list of parameterId values to drop
drop_values = ['mean_wind_direction', 'mean_radiation', 'mean_pressure', 'no_lightning_strikes', 'mean_relative_humidity', 
               'temp_soil_30', 'temp_soil_10', 'max_wind_speed_3sec', 'leaf_moisture', 'drought_index', 
               'pot_evaporation_makkink', 'no_days_acc_precip_01', 'no_days_acc_precip_1', 'vapour_pressure_deficit_mean']

# Drop rows based on the list of parameterId values
df = df[~df['parameterId'].isin(drop_values)]
```
